# Data Visualization

# Top countries with highest population

```python
df = df.sort_values(by='2021',ascending=False)
top10=df.head(10)
```

# Top  countries with highest population in 2005,2010,2015,2020

```python
plt.figure(figsize=(30,25))

plt.subplot(2,2,1)
plt.title ('Top countries with total population in 2005' , size= 20)
sns.barplot(x='2005' , y='Country Name' , data=top10 , palette = 'bright')
plt.xlabel('2022')
plt.ylabel('Country')


plt.subplot(2,2,2)
plt.title ('Top countries with total population in 2010' , size= 20)
sns.barplot(x='2010' , y='Country Name' , data=top10 , palette = 'bright')
plt.xlabel('2016')
plt.ylabel('Country')



plt.subplot(2,2,3)
plt.title ('Top countries with total population in 2015' , size= 20)
sns.barplot(x='2015' , y='Country Name' , data=top10 , palette = 'bright')
plt.xlabel('2015')
plt.ylabel('Country')



plt.subplot(2,2,4)
plt.title ('Top countries with total population in 2020' , size= 20)
sns.barplot(x='2020' , y='Country Name' , data=top10 , palette = 'bright')
plt.xlabel('2020')
plt.ylabel('Country')
plt.show()
```
